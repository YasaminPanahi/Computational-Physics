```python
import matplotlib.pyplot as plt 
import numpy as np
import timeit
```

```python
def one_dimensional_randomwalk():
    run=5000
    l=1
    P=np.linspace(0,1,20)
    t=10000
    var=[]
    mean=[]
    for p in P :
        x=[]
        for i in range (run) :
            x_walked=0
            for j in range (t) :
                a=np.random.rand()
                if a<p :
                    x_walked=x_walked+l
                else :
                    x_walked=x_walked-l
            x.append(x_walked)
        var.append(np.var(x))
        mean.append(np.mean(x))
    return  mean,var,P 
```

```python
start=timeit.default_timer()
mean,var,P=one_dimensional_randomwalk()
stop=timeit.default_timer()
print('creation time :',stop-start) 
```

```python
#plot sigma^2 vs p
start=timeit.default_timer()
fig = plt.figure(figsize=(8,6))
plt.plot(P,var,'.')
plt.title('one dimensional random walk')
plt.xlabel('probability')
plt.ylabel('variance')
z=np.polyfit(P,var,2)
#the equation is in form of y = ax^2 + bx + c (4t(p - p^2))
print('a = -4t =',z[0])
print('b = 4t =',z[1])
print('c =',z[2])
stop=timeit.default_timer()
print('creation time :',stop-start) 
```

```python
#plot <x(t)> vs p
start=timeit.default_timer()
fig = plt.figure(figsize=(8,6))
plt.plot(P,mean,'.')
plt.title('one dimensional random walk')
plt.xlabel('probability')
plt.ylabel('mean distance')
z=np.polyfit(P,mean,1)
print('gradint is 2t =',z[0])
a=[P[0],P[19]]
b=[P[0]*z[0]+z[1],P[19]*z[0]+z[1]]
plt.plot(a,b)
stop=timeit.default_timer()
print('creation time :',stop-start) 
```

```python
start=timeit.default_timer()
T = [500,550,600,650,700,750,800,850,900,950,1000] #Time
p = 0.3
run=5000
l=1
var_t=[]
mean_t=[]
for t in T :
    x_t=[]
    for i in range (run) :
        xt_walked=0
        for j in range (t) :
            a=np.random.rand()
            if a<p :
                xt_walked=xt_walked+l
            else :
                xt_walked=xt_walked-l
        x_t.append(xt_walked)
    var_t.append(np.var(x_t))
    mean_t.append(np.mean(x_t))
stop=timeit.default_timer()
print('creation time :',stop-start)     
```

```python
#plot sigma^2 vs t
start=timeit.default_timer()
fig = plt.figure(figsize=(8,6))
plt.plot(T,var_t,'.')
plt.title('one dimensional random walk for p = 0.3')
plt.xlabel('time')
plt.ylabel('variance')
z=np.polyfit(T,var_t,1)
print('gradiant =4p(1-p):',z[0])
stop=timeit.default_timer()
print('creation time :',stop-start) 
```

```python
#plot <x(t)> vs T
start=timeit.default_timer()
fig = plt.figure(figsize=(8,6))
plt.plot(T,mean_t,'.')
plt.title('one dimensional random walk for p = 0.3')
plt.xlabel('probability')
plt.ylabel('mean distance')
z=np.polyfit(T,mean_t,1)
print('gradint is 2p - 1 =',z[0])
a=[T[0],T[10]]
b=[T[0]*z[0]+z[1],T[10]*z[0]+z[1]]
plt.plot(a,b)
stop=timeit.default_timer()
print('creation time :',stop-start) 
```
