```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R=np.linspace(0,1,100000)
n=100
X=[]
```

```python
def chaos(x,r):
    return 4*r*x*(1-x)
```

```python
for r in R:
    x=np.random.rand()
    for i in range(n):
        x=chaos(x,r)   
    X.append(x)
```

```python
fig = plt.figure(figsize=(20,15))
plt.title('X final vs r')
plt.ylabel('X')
plt.xlabel('r')
plt.plot(R,X,'.')
```

```python
fig = plt.figure(figsize=(20,15))
plt.title('X final vs r')
plt.ylabel('X')
plt.xlabel('r')
plt.plot(R,X,'.')
plt.axvline(x=0.748,color = 'black')
plt.axvline(x=0.860,color = 'black')
plt.axvline(x=0.886,color = 'black')
#دهانه دومین دو شاخگی
plt.axhline(y=0.816,color = 'black')
plt.axhline(y=0.883,color = 'black')
#دهانه اولین دوشاخگی
plt.axhline(y=0.851,color = 'black')
plt.axhline(y=0.437,color = 'black')
#ّبرای آلفا
plt.axhline(y=0.5,color = 'r')
plt.axvline(x=0.81,color = 'g')
plt.axvline(x=0.875,color = 'g')
plt.axhline(y=0.38,color = 'black')
```
