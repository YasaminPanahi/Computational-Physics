```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
delta=[1000,2000,3000,4000,5000,6000,7000,8000,9000,10000,20000,30000,40000,50000,60000,70000,80000,90000,100000,200000,300000,400000,500000,600000]
t=np.float32(0.5*(10)**(-3))
```

```python
def Generate(Q,n):
    delta_t = t/delta[n] 
    C = 10**(-6) 
    R = 3*(10)**3 
    V = 10
    Q_Prime=(V/R)-(Q/(R*C))
    return np.float32(Q + delta_t*Q_Prime)
```

```python
def Generate2(Q,dt):
    C = 10**(-6) 
    R = 3*(10)**3 
    V = 10
    Q_Prime=(V/R)-(Q/(R*C))
    return Q + dt*Q_Prime
```

```python
q_T=[]
for d in delta :
    Q=np.float32(0)
    y=[np.float32(0)]
    n=delta.index(d)
    for i in range (d):
        Q=Generate(Q,n)
        y.append(Q)
    q_T.append(y[-1]) 
error=np.float32(np.abs(0.1535*(10)**(-5)-np.array(q_T)))
h=np.float32(t/np.array(delta))
```

```python
fig = plt.figure(figsize=(8,8))
plt.title('Error vs h')
plt.ylabel('log(Error = |Q_real(T) - Q_sim(T)|)')
plt.xlabel('log(h)')
plt.plot(np.log(h),np.log(error),'.')
```

```python
fig = plt.figure(figsize=(8,8))
plt.title('Error vs h')
plt.ylabel('Error = |Q_real(T) - Q_sim(T)|')
plt.xlabel('h')
plt.plot(h,error,'.')
```

```python
T=0.09
D=10
dt=T/D
x=np.linspace(0,T,D+1)
qs=[0]
q=0
for i in range (D):
    q=Generate2(q,dt)
    qs.append(q) 
```

```python
fig = plt.figure(figsize=(8,8))
plt.title('Q - t , h = 9*10**(-3)')
plt.ylabel('Q')
plt.xlabel('t')
plt.plot(x,qs,'.')
```
