```python
import numpy as np
import matplotlib.pyplot as plt
import random
import timeit
```

```python
def hoshen_kopelman_selection(a,b,c):
    if b==0 and a==0 :
        return [c,c+1,0]
    elif a!=0 and b!=0:
        key=min(a,b)
        return [key,c,1]
    else:
        key=max(a,b)
        return [key,c,2] 
```

```python
start=timeit.default_timer()
l=int(input('Please Enter Your Intended Length :'))
p=float(input('Please Enter Your Intended Probability :'))          
lattice=np.zeros((l,l+1))
for m in range(l):
      lattice[m][0]=1
color=[0,1]
counter=2
for c in range(1,l+1):
    for r in range(l):
        a=np.random.rand()
        if a<=p :
            if r!=0 :
                ans=hoshen_kopelman_selection(lattice[r-1][c],lattice[r][c-1],counter)
                lattice[r][c]=ans[0]
                counter=ans[1]
                if ans[2]==0 :
                    color.append(ans[0])
                if ans[2]==1:
                    minvalue=int(min(color[int(lattice[r-1][c])],color[int(lattice[r][c-1])]))
                    maxvalue=int(max(color[int(lattice[r-1][c])],color[int(lattice[r][c-1])]))
                    color[maxvalue]=color[minvalue]
                else:
                    if lattice[r][c-1]==0:
                        lattice[r][c]=counter
                        color.append(counter)
                        counter=counter+1
                    else:    
                        lattice[r][c]=lattice[r][c-1]

for j in range(l):
    for i in range(1,l+1):
        if lattice[j][i] == 0:
            lattice[j][i] = l*10
fig = plt.figure(figsize=(15,10))            
plt.imshow(lattice[:,1:],cmap='jet')
plt.axis('off')
stop=timeit.default_timer()
print('creation time :',stop-start)   
```
