```python
import numpy as np
import matplotlib.pyplot as plt
import random
import timeit
```

```python
def selection(a,b,c):
    if b==0 and a==0 :
        return [c,c+1,0]
    elif a!=0 and b!=0:
        key=min(a,b)
        return [key,c,1]
    else:
        key=max(a,b)
        return [key,c,0]  
```

```python
start=timeit.default_timer()
l=int(input('Please Enter Your Intended Length :'))
p=float(input('Please Enter Your Intended Probability :'))
lattice=np.zeros((l,l+1))
for i in range (l) :
    lattice[i][0]=1
c=2
for i in range(1,l+1):
    for j in range(l):
        a=np.random.rand()
        if a<p :
            if j!=0 :
                ans=selection(lattice[j-1][i],lattice[j][i-1],c)
                lattice[j][i]=ans[0]
                c=ans[1]
                if ans[2]==1 : #if both cells are on
                    k=max(lattice[j-1][i],lattice[j][i-1])
                    for ii in range(1,l+1):
                        for jj in range(l):
                            if lattice[jj][ii]==k:
                                lattice[jj][ii]=ans[0]
            else:
                if lattice[j][i-1]==0:
                    lattice[j][i]=c+1
                    c=c+1
                else:    
                    lattice[j][i]=lattice[j][i-1]
    
for j in range(l):
    for i in range(1,l+1):
        if lattice[j][i] == 0:
            lattice[j][i] = l*4 #for 'off' cells 
                   
fig = plt.figure(figsize=(15,10))
fig.set_facecolor("white")
plt.imshow(lattice[:,1:],cmap='jet')
plt.axis('off')
stop=timeit.default_timer()
print('creation time :',stop-start)  
```

```python
#check percolation (if we see number 1 in the rightest column then it's percolation,number 400 is for 'off' cells)
for i in range (l) :
    print(lattice[i][l])
```
