```python
import numpy as np
import matplotlib.pyplot as plt
import timeit
```

```python
N=5
run=10000
y=[]
counter= []
for i in range(run) :
    s=0
    for j in range(N):
        a=np.random.rand()
        s=s+a
    mean=int(1000*(s/N))
    if mean not in y:
        y.append(mean)
        counter.append(1)
    else:
        n=y.index(mean)
        counter[n]=counter[n]+1
p=[]        
for i in range(len(counter)):
    p.append(counter[i]/run)
```

```python
plt.figure(figsize=(6, 6))
plt.bar(y,p)
plt.title('N=5')
plt.xlabel('y')
plt.ylabel('abundance')
```
