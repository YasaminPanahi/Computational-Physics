```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

```python
def gaussian_distribution(sigma,n):
    y1 = np.zeros((n))
    y2 = np.zeros((n))
    for i in range(n):
        x1 = random.random()
        x2 = random.random()
        theta = float('%.2f'%(2*math.pi*x2))
        r = float('%.2f'%(sigma*np.sqrt(2*np.log((1-x1)**(-1)))))
        y1[i] = r * np.cos(theta)
        y2[i] = r * np.sin(theta)
    var_y1 = np.var(y1)
    var_y2 = np.var(y2)
    return y1,y2,var_y1,var_y2
```

```python
y1,y2,var1,var2 = gaussian_distribution(1,1000000)
var1,var2
```

```python
#plot histogram for y1 = r.cos(theta)
plt.hist(y1)
plt.xlabel('y1')
plt.ylabel('abundance')
plt.title('N = 1000000')
```

```python
#plot histogram for y2 = r.sin(theta)
plt.hist(y2)
plt.xlabel('y2')
plt.ylabel('abundance')
plt.title('N = 1000000')
```
