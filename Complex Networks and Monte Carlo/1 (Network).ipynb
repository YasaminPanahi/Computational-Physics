```python
import numpy as np
import matplotlib.pyplot as plt 
from networkx import nx
import sys
```

```python
n = 500
k1 = 0.8
k2 = 1
k3 = 8
p1 = k1/n
p2 = k2/n
p3 = k3/n
G1 = nx.erdos_renyi_graph(n,p1)
G2 = nx.erdos_renyi_graph(n,p2)
G3 = nx.erdos_renyi_graph(n,p3)
```

```python
nx.draw(G1, alpha=0.5, pos=nx.circular_layout(G1), node_size=10)
```

```python
nx.draw(G2, alpha=0.5, pos=nx.circular_layout(G2), node_size=10)
```

```python
nx.draw(G3, alpha=0.5, pos=nx.circular_layout(G3), node_size=10)
```

```python
#degree distribution and clustering
degree_distribution1 = np.zeros(n)
clustering1 = np.zeros(n)

degree_distribution2 = np.zeros(n)
clustering2 = np.zeros(n)

degree_distribution3 = np.zeros(n)
clustering3 = np.zeros(n)

for (i,j) in nx.degree(G1):
    degree_distribution1[i] = j
    
for (k,l) in nx.degree(G2):
    degree_distribution2[k] = l   

for (q,p) in nx.degree(G3):
    degree_distribution3[q] = p    
    
cluster_data1 = list(nx.clustering(G1).items())
cluster_array1 = np.array(cluster_data1)
clustering1 = cluster_array1[:,1]

cluster_data2 = list(nx.clustering(G2).items())
cluster_array2 = np.array(cluster_data2)
clustering2 = cluster_array2[:,1]

cluster_data3 = list(nx.clustering(G3).items())
cluster_array3 = np.array(cluster_data3)
clustering3 = cluster_array3[:,1]
```

```python
#plot degree distribution for <k> = 0.8
fig = plt.figure(figsize=(8,8))
plt.hist(degree_distribution1)
plt.title('degree distribution for <k> = 0.8')
plt.xlabel('k')
plt.ylabel('degree_distribution')
```

```python
#plot degree distribution for <k> = 1
fig = plt.figure(figsize=(8,8))
plt.hist(degree_distribution2)
plt.title('degree distribution for <k> = 1')
plt.xlabel('k')
plt.ylabel('degree_distribution')
```

```python
#plot degree distribution for <k> = 8
fig = plt.figure(figsize=(8,8))
plt.hist(degree_distribution3)
plt.title('degree distribution for <k> = 8')
plt.xlabel('k')
plt.ylabel('degree_distribution')
```

```python
#plot clustering for k = 0.8
fig = plt.figure(figsize=(8,8))
plt.hist(clustering1)
plt.title('clustering for <k> = 0.8')
plt.xlabel('c')
plt.ylabel('clustering')
```

```python
#plot clustering for k = 1
fig = plt.figure(figsize=(8,8))
plt.hist(clustering2)
plt.title('clustering for <k> = 1')
plt.xlabel('c')
plt.ylabel('clustering')
```

```python
#plot clustering for k = 8
fig = plt.figure(figsize=(8,8))
plt.title('clustering for <k> = 8')
plt.hist(clustering3)
plt.xlabel('c')
plt.ylabel('clustering')
```
